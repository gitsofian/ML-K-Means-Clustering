# make_blobs

```python
from sre_constants import NOT_LITERAL_UNI_IGNORE
from sklearn.datasets import make_blobs, make_moons, make_circles
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans


X, y = make_blobs(n_samples=1000, centers=5, n_features=2, random_state=0)

# Plotting data
plt.figure(1)
plt.scatter(X[:, 0], X[:, 1], c=y, marker=".")
plt.title("Datasets Truth (make blobs)")

# Erstelle K-Means
kmeans = KMeans(n_clusters=3, random_state=0, algorithm="elkan").fit(X)
y_clustering = kmeans.predict(X)

# print(f"K-Means labels Attribute: \n{kmeans.labels_}")
# print(f"K-Means Cluster Center Attribute: \n{kmeans.cluster_centers_}")


plt.figure(2)
plt.scatter(X[:, 0], X[:, 1], c=y_clustering, marker=".")
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[
            :, 1], marker="D", linewidths=2, c='#ff0000')
plt.title("Ground K-Means")

# Add Axes Labels
plt.xlabel("x1 axis")
plt.ylabel("x2 axis")

plt.show()
```

#   Elbow-Plot : make_blobs


```python
X, y = make_blobs(n_samples=1000, centers=5, n_features=2, random_state=0)

# Plotting data
plt.figure(1)
plt.scatter(X[:, 0], X[:, 1], c=y, marker=".")
plt.title("Datasets Truth (make blobs)")

# Erstelle K-Means

max_itr = 12
inertias = []
for k in range(2,max_itr):
    kmeans = KMeans(n_clusters=k, random_state=0, algorithm="elkan").fit(X)
    print(f"K-Means-Inertia : {kmeans.inertia_}")
    inertias.append(kmeans.inertia_)
    
y_clustering = kmeans.predict(X)

# print(f"K-Means labels Attribute: \n{kmeans.labels_}")
# print(f"K-Means Cluster Center Attribute: \n{kmeans.cluster_centers_}")


plt.figure(2)
plt.scatter(X[:, 0], X[:, 1], c=y_clustering, marker=".")
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[
            :, 1], marker="D", linewidths=2, c='#ff0000')
plt.title("Ground K-Means")

plt.figure(3)
plt.plot(range(2,max_itr), inertias)
plt.grid()
plt.title(f"Elbow-Plot : Inertia vs k (Anzahl Cluster) iteration")

# Add Axes Labels
plt.xlabel("x1 axis")
plt.ylabel("x2 axis")

plt.show()
```
